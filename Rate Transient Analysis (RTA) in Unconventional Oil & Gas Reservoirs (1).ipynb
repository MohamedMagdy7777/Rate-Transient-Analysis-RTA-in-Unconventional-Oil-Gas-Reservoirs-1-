```python
Matrix_Perm, TOC = df.columns[1], df.columns[4]

df.drop([Matrix_Perm, TOC], axis = 1, inplace = True)
```

```python
df.head()
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge

from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import KFold, cross_val_predict

from sklearn.pipeline import Pipeline



s = StandardScaler()
lr = LinearRegression()
lasso = Lasso()
ridge = Ridge()
kf = KFold(shuffle=True, n_splits = 3, random_state = 72018)

estimator_lr = Pipeline([("scaler", s),("regression", lr)])
estimator_lasso = Pipeline([("scaler", s),("lasso", lasso)])
estimator_ridge = Pipeline([("scaler", s),("ridge", ridge)])
```

```python
y = df['Aroot(K)']
X = df.drop(['Aroot(K)'], axis = 1)
```

```python
X
```

```python
estimator_lr.fit(X, y)
estimator_lasso.fit(X, y)
estimator_ridge.fit(X, y)
```

```python
predictions_lr = cross_val_predict(estimator_lr, X, y, cv = kf)
predictions_lasso = cross_val_predict(estimator_lasso, X, y, cv = kf)
predictions_ridge = cross_val_predict(estimator_ridge, X, y, cv = kf)
```

```python
f, ax = plt.subplots(1, 3, figsize=(20, 5))
ax[0].scatter(predictions_lr, y, c = 'blue')
ax[0].set(xlabel = 'Predictions', ylabel = 'Actual', title = 'Linear Regression')
ax[1].scatter(predictions_lasso, y, c = 'g')
ax[1].set(xlabel = 'Predictions', ylabel = 'Actual', title = 'Lasso Regression')
ax[2].scatter(predictions_ridge, y, c = 'brown')
ax[2].set(xlabel = 'Predictions', ylabel = 'Actual', title = 'Ridge Regression')
plt.show
```

```python
from sklearn.metrics import r2_score

score_lr = r2_score(predictions_lr, y)
score_lasso = r2_score(predictions_lasso, y)
score_ridge = r2_score(predictions_ridge, y)
```

```python
print(f"Linear Regression R2 Score: {round(score_lr, 5)}")
print(f"Lasso  Regression R2 Score: {round(score_lasso, 5)}")
print(f"Ridge  Regression R2 Score: {round(score_ridge, 5)}")
```

## Polynomial Features

```python
from sklearn.preprocessing import PolynomialFeatures

pf = PolynomialFeatures(degree=2)

estimator_pflinear = Pipeline([("polynomial_features", pf),("scaler", s),("linear_regression", lr)])
estimator_pflasso = Pipeline([("polynomial_features", pf),("scaler", s),("lasso_regression", lasso)])
estimator_pfridge = Pipeline([("polynomial_features", pf),("scaler", s),("ridge_regression", ridge)])
estimator_pflinear.fit(X, y)
estimator_pflasso.fit(X, y)
estimator_pfridge.fit(X, y)
```

```python
predictions_pflinear = cross_val_predict(estimator_pflinear, X, y, cv = kf)
predictions_pflasso = cross_val_predict(estimator_pflasso, X, y, cv = kf)
predictions_pfridge = cross_val_predict(estimator_pfridge, X, y, cv = kf)
```

```python
f, ax = plt.subplots(1, 3, figsize=(20, 5))
ax[0].scatter(predictions_pflinear, y, c = 'blue')
ax[0].set(xlabel = 'Predictions', ylabel = 'Actual', title = 'Linear Regression with Polynomial Feature (2 degrees)')
ax[1].scatter(predictions_pflasso, y, c = 'g')
ax[1].set(xlabel = 'Predictions', ylabel = 'Actual', title = 'Lasso Regression with Polynomial Feature (2 degrees)')
ax[2].scatter(predictions_pfridge, y, c = 'brown')
ax[2].set(xlabel = 'Predictions', ylabel = 'Actual', title = 'Ridge Regression with Polynomial Feature (2 degrees)')
plt.show
```

```python
score_pflinear = r2_score(predictions_pflinear, y)
score_pflasso = r2_score(predictions_pflasso, y)
score_pfridge = r2_score(predictions_pfridge, y)
```

```python
print(f"Linear Regression with Polynomial Features (degree = 2) R2 Score: {round(score_pflinear, 5)}")
print(f"Lasso  Regression with Polynomial Features (degree = 2) R2 Score: {round(score_pflasso, 5)}")
print(f"Ridge  Regression with Polynomial Features (degree = 2) R2 Score: {round(score_pfridge, 5)}")
```
